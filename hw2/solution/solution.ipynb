```python
import numpy as np
from soft_svm import SoftSVM
import matplotlib.pyplot as plt
from data import get_data, prepare_data
```

```python
# create blob
train_data, test_data = get_data()
X_train, y_train = prepare_data(train_data)
X_test, y_test = prepare_data(test_data)

# scatter plot
data = train_data
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='viridis')
plt.xlim(-16,8)
plt.ylim(-15,15)
plt.show()
plt.close()
```

```python
clf_full = SoftSVM(2)
clf_full.train(X_train=X_train, y_train=y_train,
          X_test=X_test, y_test=y_test,
          n_iterations=1000, batch_size=10000,
          display_step=1, learning_rate=0.1, random_seed=10)

clf_mini = SoftSVM(2)
clf_mini.train(X_train=X_train, y_train=y_train,
          X_test=X_test, y_test=y_test,
          n_iterations=1000, batch_size=128,
          display_step=1, learning_rate=0.01, random_seed=10)

clf_stochastic = SoftSVM(2)
clf_stochastic.train(X_train=X_train, y_train=y_train,
                     X_test=X_test, y_test=y_test,
                     n_iterations=1000, batch_size=1,
                     display_step=1, learning_rate=0.001, random_seed=10)
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))
axs[0].plot(clf_full.trajectories['train_loss'], label='train_loss')
axs[0].plot(clf_full.trajectories['test_loss'], label='test_loss')
axs[0].set_xlabel('Iterations')
axs[0].set_ylabel('Loss (mean)')
axs[0].legend()

axs[1].plot(clf_mini.trajectories['train_loss'], label='train_loss')
axs[1].plot(clf_mini.trajectories['test_loss'], label='test_loss')
axs[1].set_xlabel('Iterations')
axs[1].set_ylabel('Loss (mean)')
axs[1].legend()

axs[2].plot(clf_stochastic.trajectories['train_loss'], label='train_loss')
axs[2].plot(clf_stochastic.trajectories['test_loss'], label='test_loss')
axs[2].set_xlabel('Iterations')
axs[2].set_ylabel('Loss (mean)')
axs[2].legend()

plt.savefig('g.png')
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))
axs[0].plot(clf_full.trajectories['time'], clf_full.trajectories['train_loss'], label='train_loss')
axs[0].plot(clf_full.trajectories['time'], clf_full.trajectories['test_loss'], label='test_loss')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Loss (mean)')
axs[0].legend()

axs[1].plot(clf_mini.trajectories['time'], clf_mini.trajectories['train_loss'], label='train_loss')
axs[1].plot(clf_mini.trajectories['time'], clf_mini.trajectories['test_loss'], label='test_loss')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Loss (mean)')
axs[1].legend()

axs[2].plot(clf_stochastic.trajectories['time'], clf_stochastic.trajectories['train_loss'], label='train_loss')
axs[2].plot(clf_stochastic.trajectories['time'], clf_stochastic.trajectories['test_loss'], label='test_loss')
axs[2].set_xlabel('Time (s)')
axs[2].set_ylabel('Loss (mean)')
axs[2].legend()

plt.savefig('h.png')
```

```python
fig, axs = plt.subplots(1,3, figsize=(20,5))
axs[0].plot(clf_full.trajectories['train_accuracy'], label='train_loss')
axs[0].plot(clf_full.trajectories['test_accuracy'], label='test_loss')
axs[0].set_xlabel('Iterations')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

axs[1].plot(clf_mini.trajectories['train_accuracy'], label='train_loss')
axs[1].plot(clf_mini.trajectories['test_accuracy'], label='test_loss')
axs[1].set_xlabel('Iterations')
axs[1].set_ylabel('Accuracy')
axs[1].legend()

axs[2].plot(clf_stochastic.trajectories['train_accuracy'], label='train_loss')
axs[2].plot(clf_stochastic.trajectories['test_accuracy'], label='test_loss')
axs[2].set_xlabel('Iterations')
axs[2].set_ylabel('Accuracy')
axs[2].legend()

plt.savefig('i.png')
```
